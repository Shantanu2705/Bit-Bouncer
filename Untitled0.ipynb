```python
1) intialize the open list
2) initialize the close list put the starting node in the open list
3) while the open list is not empty
  a) find the node with least f and name it as q
  b) pop q off the list
  c) generate q's 8 successors and set their parents as q
  d) for each successor
    i) if the successor is the goal, stop search
    ii) else, compute for both g and h as successor
      successor.g= q.g + distance between successor and q
      successor.h= distance between goal and successor
      successor.f= successor.g+successor.h
    iii) if the node is in same position as successor in the open list which has lower f than successor, skip that successor
    iv) if the node is in same position as successor as in the closed list which has lower f than successor, skip that successor
      otherwise, put the node in the open list
  end(for loop)
  e) put q in the closed list
end(while loop)
```
